## Let's read the json file

## keep only columns that include venue name, and anything that is associated with location

```python
filtered_columns = ['name'] + [col for col in dataframe.columns if col.startswith('location.')] + ['id']
dataframe_filtered = dataframe.loc[:, filtered_columns]

dataframe_filtered.columns = [column.split('.')[-1] for column in dataframe_filtered.columns]

dataframe_filtered
```

```python
dataframe_filtered.shape
```

## Let's Create a map with neighborhood Pressure Washing businesss using Folium

```python
Neighborhood_map = folium.Map(location=[latitude, longitude], zoom_start=9) # generate map centred around the home.

# add a red circle marker to represent the HOME
folium.CircleMarker(
    [latitude, longitude],
    radius=10,
    color='red',
    popup='HOME',
    fill = True,
    fill_color = 'red',
    fill_opacity = 0.6
).add_to(Neighborhood_map)


# add markers to map

for lat, lng, label in zip(dataframe_filtered['lat'], dataframe_filtered['lng'], dataframe_filtered['name']):
    label = folium.Popup(label, parse_html=True)
    folium.CircleMarker(
        [lat, lng],
        radius=5,
        popup=label,
        color='blue',
        fill=True,
        fill_color='#3186cc',
        fill_opacity=0.7,
        parse_html=False).add_to(Neighborhood_map)  
    
Neighborhood_map
```

## Next, let's group rows by and by taking the mean of the frequency of occurrence of each category


```python
Neighborhoods = dataframe_filtered.groupby(['city','name','postalCode']).mean().reset_index()
print(Neighborhoods.shape)
Neighborhoods
```

```python
Neighborhoods_map = folium.Map(location=[latitude, longitude], zoom_start=9) # generate map centred around the home.

# add a red circle marker to represent the HOME
folium.CircleMarker(
    [latitude, longitude],
    radius=10,
    color='red',
    popup='HOME',
    fill = True,
    fill_color = 'red',
    fill_opacity = 0.6
).add_to(Neighborhoods_map)


# add markers to map

for lat, lng, label in zip(Neighborhoods['lat'], Neighborhoods['lng'], Neighborhoods['name']):
    label = folium.Popup(label, parse_html=True)
    folium.CircleMarker(
        [lat, lng],
        radius=5,
        popup=label,
        color='blue',
        fill=True,
        fill_color='#3186cc',
        fill_opacity=0.7,
        parse_html=False).add_to(Neighborhoods_map)  
    
Neighborhoods_map
```

## Cluster Neighborhoods

```python
Neighborhoods_group = dataframe_filtered.groupby(['postalCode']).mean().reset_index()

kclusters =4

Neighborhood_clustering = Neighborhoods_group.drop('postalCode',1)
kmeans = KMeans(n_clusters=kclusters, random_state=0).fit(Neighborhood_clustering)
kmeans.labels_[0:10] 
Neighborhood_clustering.insert(0,'Cluster',kmeans.labels_)
Neighborhood_clustering
```

```python
map_clusters = folium.Map(location=[latitude, longitude], zoom_start=10)

# set color scheme for the clusters
x = np.arange(kclusters)
ys = [i+x+(i*x)**2 for i in range(kclusters)]
colors_array = cm.rainbow(np.linspace(0, 1, len(ys)))
rainbow = [colors.rgb2hex(i) for i in colors_array]

# add markers to the map
markers_colors = []
for lat, lon, post, city, name, cluster in zip(Neighborhoods['lat'], Neighborhoods['lng'], Neighborhoods['postalCode'], Neighborhoods['city'], Neighborhoods['name'], Neighborhood_clustering['Cluster']):
    label = folium.Popup('{} ({}): {} - Cluster {}'.format(post, city, name, cluster), parse_html=True)
    folium.CircleMarker(
        [lat, lon],
        radius=5,
        popup=label,
        color=rainbow[cluster-1],
        fill=True,
        fill_color=rainbow[cluster-1],
        fill_opacity=0.7).add_to(map_clusters)
       
map_clusters
```
